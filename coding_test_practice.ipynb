【課題 1】「合格者」の平均点を出せ
リスト A には、あるクラスのテストの点数が入っています。 「60点以上」の人だけ の 平均点 を計算して返してください。 （※ 60点未満の人は無視します）

```python
def solution(A):
    #60点以上の人の点数の合計
    s = 0
    #60点以上の人の人数
    count = 0
    #「60点以上」の人だけ の 平均点
    ave = 0
    for x in A:
        if x >= 60:
            s += x
            count += 1
    if count == 0:
      return 0
    else:
      ave = s // count
      return ave
        


# --- 確認用 ---
if __name__ == "__main__":
    print(solution([80, 40, 60, 90, 50])) 
```

【課題 2】足して 100 になるペアを探せ
問題内容： 数字のリスト A が渡されます。そのリストの中に、足すとちょうど 100 になる「2つの異なる要素」の組み合わせがあるかどうかを判定してください。

存在すれば True を返す。

存在しなければ False を返す。

注意点：

リストの中に 50 が1つしかない場合、50 + 50 = 100 と判定してはいけません。あくまで「別の場所にある2つの数字」を足す必要があります。

```python
def solution(A):
    
    for x in A:
        
        for y in A:
            if x == y:
              continue
            elif x + y == 100:
              return True
    return False

# --- 動作確認用 ---
if __name__ == "__main__":
    # ケース1: 40 + 60 = 100 なので True
    print(f"Test 1: {solution([10, 40, 50, 60])}") 
    
    # ケース2: 100になるペアはないので False
    print(f"Test 2: {solution([10, 20, 30])}")     
    
    # ケース3: 50が1つしかないのでペアは作れず False
    print(f"Test 3: {solution([50, 10, 20])}")
```

このままでは、リストに同じ数字が入っていた場合に見逃してしまいます。そのためrangeで何番目の中身なのかを区別して、見逃しを防ぐようにしました。

```python
def solution(A):
    
    for i in range(len(A)):
      x = A[i]
      for j in range(len(A)):
        y = A[j]      
        if i == j:
          continue
        elif x + y == 100:
          return True
    return False

# --- 動作確認用 ---
if __name__ == "__main__":
    # ケース1: 40 + 60 = 100 なので True
    print(f"Test 1: {solution([10, 40, 50, 60])}") 
    
    # ケース2: 100になるペアはないので False
    print(f"Test 2: {solution([10, 20, 30])}")     
    
    # ケース3: 50が1つしかないのでペアは作れず False
    print(f"Test 3: {solution([50, 10, 20])}")
```
